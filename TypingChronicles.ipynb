```python
def Progress():
    def H():
        ax.clear()
        f1=open("progress3.txt","r")
        string1=f1.read()
        ls1=string1.split("\n")
        ls2=[]
        for x in ls1:
            ls2.append(int(x))
        ls1=np.array(ls2)
        ax.set_title("Let's Test Your Typing")
        ax.plot(range(1,len(ls2)+1), ls2, color = '#b2d8d8',linestyle = 'solid', marker = 'o', markersize = 2)
        canvas.draw()
    def M():
        ax.clear()
        f1=open("progress2.txt","r")
        string1=f1.read()
        ls1=string1.split("\n")
        ls2=[]
        for x in ls1:
            ls2.append(int(x))
        ls1=np.array(ls2)
        ax.set_title("Practice Sentences")
        ax.plot(range(1,len(ls2)+1), ls2, color = '#ffd68a',linestyle = 'solid', marker = 'o', markersize = 2)
        canvas.draw()
    def E():
        ax.clear()
        f1=open("progress1.txt","r")
        string1=f1.read()
        ls1=string1.split("\n")
        ls2=[]
        for x in ls1:
            ls2.append(int(x))
        ls1=np.array(ls2)
        ax.set_title("Practice Words")
        ax.plot(range(1,len(ls2)+1), ls2, color = '#695E93',linestyle = 'solid', marker = 'o', markersize = 2)
        canvas.draw()
    root = tk.Tk()
    fig, ax = plt.subplots()

    # Tkinter Application
    frame = tk.Frame(root)


    # Create Canvas
    canvas = FigureCanvasTkAgg(fig, master=root)  
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)
    frame.pack()
    tk.Button(frame,text="Let's Test Your Typing",command=H).pack(pady=10)
    tk.Button(frame,text="Practice Sentences",command=M).pack(pady=10)
    tk.Button(frame,text="Practice Words",command=E).pack(pady=10)

    root.mainloop()
```
